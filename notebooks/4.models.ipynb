# RandomForestRegressor model.

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the model (using a simple default for example)
rfr_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rfr_model.fit(x_train, y_train_p) # Note: X_train must be the scaled/encoded version

y_pred_p=rfr_model.predict(x_test)
```

```python

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

r2 = r2_score(y_test_p, y_pred_p)


mae = mean_absolute_error(y_test_p, y_pred_p)


mse = mean_squared_error(y_test_p, y_pred_p)
rmse = np.sqrt(mse)

print("--- Model Accuracy for Protein Feed Yield ---")
print(f"R-squared (R²) Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f} percentage points")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f} percentage points")
```

# K-Fold Cross-Validation

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# --- K-Fold setup ---
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# --- Model ---
rfr_model = RandomForestRegressor(n_estimators=100, random_state=42)

# --- Custom scorers ---
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)
rmse_scorer = make_scorer(lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, y_pred)), greater_is_better=False)

# --- Cross-validation ---
r2_scores = cross_val_score(rfr_model, x_train, y_train_p, cv=kf, scoring='r2')
mae_scores = cross_val_score(rfr_model, x_train, y_train_p, cv=kf, scoring=mae_scorer)
rmse_scores = cross_val_score(rfr_model, x_train, y_train_p, cv=kf, scoring=rmse_scorer)

# --- Results ---
print("🔹 R² mean: {:.4f} | std: {:.4f}".format(np.mean(r2_scores), np.std(r2_scores)))
print("🔹 MAE mean: {:.4f} | std: {:.4f}".format(-np.mean(mae_scores), np.std(mae_scores)))
print("🔹 RMSE mean: {:.4f} | std: {:.4f}".format(-np.mean(rmse_scores), np.std(rmse_scores)))
```

# Result : R² mean = 0.7462:  model explains about 74.6% of the variance in protein yield.

# Grid search CV ( optimize the hyperparameters ) :

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# --- Base model ---
rfr_model = RandomForestRegressor(random_state=42)

# --- Parameter grid (tweak values if you want deeper search) ---
param_grid = {
    'n_estimators': [100, 200, 300],         # number of trees
    'max_depth': [None, 10, 20, 30],         # depth of each tree
    'min_samples_split': [2, 5, 10],         # min samples to split an internal node
    'min_samples_leaf': [1, 2, 4],           # min samples per leaf node
    'max_features': ['auto', 'sqrt']         # how many features to consider when splitting
}

# --- Grid Search ---
grid_search = GridSearchCV(
    estimator=rfr_model,
    param_grid=param_grid,
    cv=5,                    # 5-fold cross-validation
    scoring='r2',            # optimize for R²
    n_jobs=-1,               # use all CPU cores
    verbose=2                # print progress
)

# --- Fit the grid search ---
grid_search.fit(x_train, y_train_p)

# --- Best parameters and score ---
print("🌟 Best Parameters:", grid_search.best_params_)
print("🌟 Best R² Score:", grid_search.best_score_)

# --- Best model ---
best_model = grid_search.best_estimator_

# --- Evaluate on test set ---
y_pred_best = best_model.predict(x_test)
r2_best = r2_score(y_test_p, y_pred_best)
mae_best = mean_absolute_error(y_test_p, y_pred_best)
rmse_best = np.sqrt(mean_squared_error(y_test_p, y_pred_best))

print("\n🔹 Test Set Performance:")
print(f"R²: {r2_best:.4f}")
print(f"MAE: {mae_best:.4f}")
print(f"RMSE: {rmse_best:.4f}")
```

# WHY RANDOM FOREST ?


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# --- 1. Define Target Variables for Wheat ---
# (Making sure these are available from your previous cells)
try:
    y_train_w = y_train[['wheat']]
    y_test_w = y_test[['wheat']]
except NameError:
    print("Error: 'y_train' or 'y_test' not found. Please re-run cells 8 and 11.")
    # You could also stop execution here if this were a script

# --- 2. Gradient Boosting Regressor (GBR) ---
print("--- Starting Gradient Boosting Regressor (GBR) ---")

# Define a simple parameter grid for GBR
param_grid_gbr = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5, 10]
}

# --- Grid Search for GBR ---
grid_search_gbr = GridSearchCV(
    estimator=GradientBoostingRegressor(random_state=42),
    param_grid=param_grid_gbr,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=1
)

print("Fitting GBR model...")
grid_search_gbr.fit(x_train, y_train_w.values.ravel())

# --- GBR Results ---
best_model_gbr = grid_search_gbr.best_estimator_
y_pred_gbr = best_model_gbr.predict(x_test)

r2_gbr = r2_score(y_test_w, y_pred_gbr)
mae_gbr = mean_absolute_error(y_test_w, y_pred_gbr)
rmse_gbr = np.sqrt(mean_squared_error(y_test_w, y_pred_gbr))

print("\n✅ GBR Test Set Performance:")
print(f"Best GBR R² (CV): {grid_search_gbr.best_score_:.4f}")
print(f"Best GBR Params: {grid_search_gbr.best_params_}")
print(f"Test R²: {r2_gbr:.4f}")
print(f"Test MAE: {mae_gbr:.4f}")
print(f"Test RMSE: {rmse_gbr:.4f}")
print("-" * 40)


# --- 3. Support Vector Regressor (SVR) ---
print("\n--- Starting Support Vector Regressor (SVR) ---")

# Create a Pipeline: 1) Scale data, 2) Run SVR
# SVR is very sensitive to feature scaling!
svr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR())
])

# Define a parameter grid for SVR
# Note: We prefix params with 'svr__' to tell the pipeline
# which step they belong to.
param_grid_svr = {
    'svr__kernel': ['linear', 'rbf'],
    'svr__C': [1, 10, 50],
    'svr__gamma': ['scale', 'auto']
}

# --- Grid Search for SVR Pipeline ---
grid_search_svr = GridSearchCV(
    estimator=svr_pipeline,
    param_grid=param_grid_svr,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=1
)

print("Fitting SVR model (with scaling)...")
# We fit on the *original* x_train. The pipeline handles scaling.
grid_search_svr.fit(x_train, y_train_w.values.ravel())

# --- SVR Results ---
best_model_svr = grid_search_svr.best_estimator_
y_pred_svr = best_model_svr.predict(x_test)

r2_svr = r2_score(y_test_w, y_pred_svr)
mae_svr = mean_absolute_error(y_test_w, y_pred_svr)
rmse_svr = np.sqrt(mean_squared_error(y_test_w, y_pred_svr))

print("\n✅ SVR Test Set Performance:")
print(f"Best SVR R² (CV): {grid_search_svr.best_score_:.4f}")
print(f"Best SVR Params: {grid_search_svr.best_params_}")
print(f"Test R²: {r2_svr:.4f}")
print(f"Test MAE: {mae_svr:.4f}")
print(f"Test RMSE: {rmse_svr:.4f}")
print("-" * 40)
```
