Merging tables


```python
df=data
df
```

```python
df = pd.merge(
    df,
    region,
    on='File Code',  # The common column
    how='left'          
)

df
```

```python
df.drop(columns=['BLS Region 2_x','BLS Region 1','BLS Code_y','BLS Region 2_y'],inplace=True)
```

```python
r={'BLS Code_x':'BLS Code'}
df.rename(columns=r,inplace=True)
df
```

```python

import pandas as pd
# Assuming df_master is your merged and cleaned DataFrame

# Define the file name for your new Excel sheet
file_name = 'final.xlsx'

# 1. Export the DataFrame to an Excel file
df.to_excel(
    file_name,
    index=False, # IMPORTANT: This prevents writing the DataFrame's row numbers as a column in the Excel file
    sheet_name='Prediction_Data'
)

print(f"Successfully saved your merged table to: {file_name}")
```
