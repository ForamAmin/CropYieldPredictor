```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a=le.fit_transform(data['CO2 effects'])
```

```python
data['CO2 effects']=a
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
Scenario=ohe.fit_transform(data[['Scenario']]).toarray()
```

```python
new_col_names = ohe.get_feature_names_out(['Scenario'])
scenario_encoded=pd.DataFrame(Scenario,columns=new_col_names,index=data.index)
```

```python
data=pd.concat([data,scenario_encoded],axis=1)
data
```

```python
data.head(20)
```

```python
data.drop(columns=['Scenario'])
```

```python
# 1. Define the desired order explicitly
adaptation_order = ['No', 'Level 1', 'Level 2']

# 2. Convert the column to the Categorical data type with the defined order
data['Adapt- ation'] = pd.Categorical(
    data['Adapt- ation'], 
    categories=adaptation_order, 
    ordered=True
)


```

```python
data.head(20)
```

```python
# 3. Apply the Label Encoding (converting the categorical order to integers)
data['Adaptation'] = data['Adapt- ation'].cat.codes
```

```python
data.drop(columns='Adapt- ation',inplace=True)
```

```python
data.head(20)
```

```python
import numpy as np
data['Time_Slice']=data['Time_Slice'].replace('Equilibrium',2150.0)

```

```python
data
```

```python
data.drop(columns=['Scenario','Scenario Code'],inplace=True)
```

```python
ohe2=OneHotEncoder()
bls=ohe2.fit_transform(data[['BLS Code']]).toarray()
bls
```

```python
new_col = ohe2.get_feature_names_out(['BLS Code'])
bls_encoded=pd.DataFrame(bls,columns=new_col,index=data.index)
data=pd.concat([data,bls_encoded],axis=1)
```

```python
data
```

```python

from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
ms=m.fit_transform(data[['Time_Slice']])
```

```python
data['Time_Slice']=ms
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
ss=s.fit_transform(data[['CO2 ppm']])
```

```python
data['CO2 ppm']=ss
```

```python
data.drop(columns=['BLS Region','File Code'],inplace=True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

# 1. Define the columns to be excluded from the features (X)
target_columns = ['wheat', 'rice', 'coarse grains', 'protein feed', 'grains', 'four commo-dities']

# 2. Define Features (X)
# Drop the target columns from the full dataset (data)
X = data.drop(columns=target_columns, axis=1)

# 3. Define Targets (Y)
# Select ALL target columns by passing the list (target_columns) to data[]
Y = data[target_columns]

# 4. Perform the Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(
    X,
    Y,
    test_size=0.2, # It's good practice to explicitly state the test size
    random_state=42
)

print("Data split successfully!")
print(f"X_train shape: {x_train.shape}")
print(f"Y_train shape: {y_train.shape}")
```
